```python
from lib import Preprocessor, Dense_Keras
```

```python
uri = 'https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv'
nn_data = Preprocessor(uri)
```

```python
nn_data.binning('application_type', 500)
nn_data.binning('classification', 500)
```

```python
drop_ids = ['name', 'ein']
nn_data.cleanup(drop_ids)
ml_data = nn_data.split_data('is_successful', random_state = 78)
```

```python
layers = []
layers.append({'units': 18, 'activation': 'relu'})
layers.append({'units': 12, 'activation': 'relu'})
layers.append({'units': 12, 'activation': 'relu'})
nn_model = Dense_Keras(ml_data, layers)
nn_model.compile()
nn_model.train(epochs = 20)
```

```python
config = nn_model.model.get_config()
```

```python
for i in config:
  print(i)
```

```python
layer_configs = config.get('layers')
```

```python
unit_list = []
act_list = []
for layer in layer_configs:
  layer: dict
  c = layer.get('config')
  try:
    units = c['units']
  except:
    units = c.get('batch_shape')[1]
  finally:
    unit_list.append(units)
  act_list.append(c.get('activation'))
```
